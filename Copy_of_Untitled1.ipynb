```python
#QUES 9
import random
import string

def generate_random_string(length):
    letters = string.ascii_lowercase
    return ''.join(random.choice(letters) for _ in range(length))

# Open the file in write mode
with open("random_strings.txt", "w") as file:
    # Generate and write 100 random strings
    for _ in range(100):
        random_length = random.randint(10, 15)
        random_string = generate_random_string(random_length)
        file.write(random_string + "\n")
```

```python
#QUES 10
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True

# Open the file in write mode
with open("prime_numbers.txt", "w") as file:
    # Find and write prime numbers between 600 and 800
    for num in range(600, 801):
        if is_prime(num):
            file.write(str(num) + "\n")
```

```python
#QUES 12
import time
import matplotlib.pyplot as plt

def sort_list(num_elements):
    lst = list(range(num_elements, 0, -1))  # Generate a list in reverse order

    start_time = time.time()
    sorted_lst = sorted(lst)
    end_time = time.time()

    return end_time - start_time

# Define the number of elements in the list
num_elements_list = [5000, 10000, 15000, 20000, 25000]

# Measure the time taken for each number of elements
time_taken_list = []
for num_elements in num_elements_list:
    time_taken = sort_list(num_elements)
    time_taken_list.append(time_taken)

# Plot the graph
plt.plot(num_elements_list, time_taken_list, marker='o')
plt.xlabel('Number of Elements')
plt.ylabel('Time Taken (seconds)')
plt.title('Sorting Time Taken vs Number of Elements')
plt.grid(True)
plt.show()
```

```python
#QUES 13
def calculate_average(marks):
    total_marks = sum(marks)
    num_subjects = len(marks)
    average = total_marks / num_subjects
    return average

# Dictionary of student marks
student_marks = {
    "Alice": [85, 90, 92, 88, 95],
    "Bob": [78, 82, 80, 75, 85],
    "Charlie": [90, 92, 88, 86, 89],
    "David": [82, 80, 75, 85, 78],
    "Eve": [92, 88, 86, 89, 90]
}

# Calculate average marks for each student
average_marks = {}
for student, marks in student_marks.items():
    average_marks[student] = calculate_average(marks)

# Find student with maximum average marks
max_student = max(average_marks, key=average_marks.get)
max_average = average_marks[max_student]

# Find student with minimum average marks
min_student = min(average_marks, key=average_marks.get)
min_average = average_marks[min_student]

# Print the results
print("Student with maximum average marks:")
print("Name:", max_student)
print("Average Marks:", max_average)
print()
print("Student with minimum average marks:")
print("Name:", min_student)
print("Average Marks:", min_average)
```
